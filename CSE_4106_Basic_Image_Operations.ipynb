# ***Reading & Displaying an image***
* OpenCV-Python is a library of Python bindings designed to solve computer vision problems.
* cv2.imread() method loads an image from the specified file.
* If the image cannot be read (because of missing file, improper permissions, unsupported or invalid format) then this method returns an empty matrix.
* Syntax: cv2.imread(path, flag)

* Parameters:
* path: A string representing the path of the image to be read.
* flag: It specifies the way in which image should be read. Itâ€™s default value is cv2.IMREAD_COLOR.
1.   cv2.IMREAD_COLOR or 1
2.   cv2.IMREAD_GRAYSCALE 0r 0
3.   cv2.IMREAD_UNCHANGED or -1
* Return Value: This method returns an image that is loaded from the specified file.


```python
#import the libraries
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
#reading the image
image = cv2.imread('/content/rgb.jpg', 1)
cv2_imshow(image)
plt.imshow(image)
#save the image
cv2.imwrite('testing_save.png',image)
```

# ***Changing the color spaces***


*   A color space is a protocol for representing colors in a way that makes them easily reproducible. We know that grayscale images have single pixel values and color images contain 3 values for each pixel the intensities of the Red, Green and Blue channels.
*  OpenCV reads a given image in the BGR format by default. So, youâ€™ll need to change the color space of your image from BGR to RGB when reading images using OpenCV.
* For color conversion, we use the function cv2.cvtColor(input image, flag) where flag deter- mines the type of conversion. 

1. For BGR ->Gray conversion we use the flags cv2.COLOR_BGR2GRAY.
2. For BGR -> RGB conversion we use the flags cv2.COLOR_BGR2RGB.
3. For BGR -> HSV, we use the flag cv2.COLOR_BGR2HSV. 


**Convert to Gray Scale**

```python
#converting the image into grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#display the image
cv2_imshow(image_gray)
plt.imshow(image_gray)
```

```python
#converting the image in HSV color space
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
#display the image
cv2_imshow(image_hsv)
plt.imshow(image_hsv)
```

```python
#converting the image in RGB color space
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#display the image
cv2_imshow(image_rgb)
plt.imshow(image_rgb)
```

# ***Accessing Image Properties***
* Properties of the image such as shape and size are accessed using NumPy array functions.
* Image properties include number of rows, columns and channels, type of image data, number of pixels etc.

# ***Splitting & Merging Image Channels***
* The B,G,R channels of an image can be split into their individual planes when needed.
* Then, the individual channels can be merged back together to form a BGR image again. 
* In OpenCV, an image is represented as a NumPy array of shape (height, width, num_channels), where num_channels corresponds to the number of color channels in the image. For a color image, the num_channels is typically 3, corresponding to the Red, Green, and Blue channels (RGB).
* Splits the image array into separate pixel arrays for each color channel (Red, Green, and Blue) using the split () function, and stores them in the r, g, and b variables.
* Merges the individual color channels into a single RGB image using the merge () function.

```python
#Splitting Image Channels
b, g, r = cv2.split(image)
rgb_image = cv2.merge((r, g, b))
num_red_pixels = cv2.countNonZero(r)
num_green_pixels = cv2.countNonZero(g)
num_blue_pixels = cv2.countNonZero(b)
# Display the total number of pixels for each color channel
print('Total number of red pixels:', num_red_pixels)
print('Total number of green pixels:', num_green_pixels)
print('Total number of blue pixels:', num_blue_pixels)
```

```python
#Displaying different image channels
#creates a figure with a size of 12 inches by 10 inches and adds two subplots to it.
fig = plt.figure(figsize=(12,10))
# Creating subplot for top left corner of the grid
ax1 = fig.add_subplot(221)
# Creating subplot for top right corner of the grid 
ax2 = fig.add_subplot(222) 
ax1.imshow(image)
ax1.title.set_text('BGR Channeled Image')
ax2.imshow(rgb_image)
ax2.title.set_text('RGB Channeled Image')
```

# ***Converting Image From One Format to Another***

```python
import os
import cv2

# Set the directory path containing JPEG images
input_dir = '/content/drive/MyDrive/Jpg Images'

# Create a directory for storing the output PNG images
output_dir = '/content/drive/MyDrive/PNG Images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all the JPEG images in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Load the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Convert the image to PNG format
        png_filename = os.path.splitext(filename)[0] + '.png'
        png_path = os.path.join(output_dir, png_filename)
        cv2.imwrite(png_path, img)
```
