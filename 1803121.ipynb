```python
#import the libraries
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
#reading the image
image = cv2.imread('/content/rgb.jpg', 1)
cv2_imshow(image)
plt.imshow(image)
#save the image
cv2.imwrite('testing_save.png',image)
```

**cvtColor:** It is used to convert between colors format. Here it is used to convert BGR format to RGB

```python
#Splitting Image Channels
b, g, r = cv2.split(image)
rgb_image = cv2.merge((r, g, b))
num_red_pixels = cv2.countNonZero(r)
num_green_pixels = cv2.countNonZero(g)
num_blue_pixels = cv2.countNonZero(b)
# Display the total number of pixels for each color channel
print('Total number of red pixels:', num_red_pixels)
print('Total number of green pixels:', num_green_pixels)
print('Total number of blue pixels:', num_blue_pixels)
```

**split** : It is used to split multi-channel image into separate single-channel images. This mathod takes a multi-channel image as input and returns a tuple of single-channel images, each corresponding to one of the channels in the original image.

**merge** : It is used to merge multiple single-channel images into a single multi-channel image.This mathod takes a tuple of single-channel images as input and returns a multi-channel image where each channel corresponds to one of the input images.

**countNonZero** : It is used to count the number of non-zero pixels in a single-channel image.This mathod takes a single-channel image as input and returns the number of non-zero pixels in the image. A non-zero pixel is any pixel in the image whose intensity value is greater than zero.


```python
#Displaying different image channels
#creates a figure with a size of 12 inches by 10 inches and adds two subplots to it.
fig = plt.figure(figsize=(12,10))
# Creating subplot for top left corner of the grid
ax1 = fig.add_subplot(221)
# Creating subplot for top right corner of the grid 
ax2 = fig.add_subplot(222) 
ax1.imshow(image)
ax1.title.set_text('BGR Channeled Image')
ax2.imshow(rgb_image)
ax2.title.set_text('RGB Channeled Image')
```

**figrure(figsize(width, height))** : This mathod is used to set the size of a figure. It takes a tuple of (width, height) in inches as input, where width is the width of the figure and height is the height of the figure.

**add_subplot()** : This method is used to add a subplot to a figure with a specified size. It takes three arguments: nrows, ncols, and index. These arguments specify the number of rows and columns in the subplot grid and the index of the current subplot. Here it only take one paratmeter index.

**imshow()** : It is used to display an image in a window. The function takes two arguments: the name of the window and the image to be displayed.




```python
fig, axs = plt.subplots(1, 3, figsize=(10, 5))
# Display the red channel image in the first subplot
axs[0].imshow(r)
axs[0].set_title('Red Channel')
# Display the green channel image in the second subplot
axs[1].imshow(g)
axs[1].set_title('Green Channel')
# Display the blue channel image in the third subplot
axs[2].imshow(b)
axs[2].set_title('Blue Channel')
```

```python
# Extract the channel pixel array
r = image[:,:,2]
g = image[:,:,1]
b = image[:,:,0]
print(r)
```

Here the code extracts the pixel arrays for the red, green, and blue channels of the image using NumPy array slicing. Specifically, it uses the [:,:,2], [:,:,1], and [:,:,0] indexing expressions to extract the third, second, and first channels of the image, respectively. These channels correspond to the red, green, and blue color channels of the image.

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 5))
# Display the red channel image in the first subplot
axs[0].imshow(r, cmap='gray')
axs[0].set_title('Red Channel')
# Display the green channel image in the second subplot
axs[1].imshow(g, cmap='gray')
axs[1].set_title('Green Channel')
# Display the blue channel image in the third subplot
axs[2].imshow(b, cmap='gray')
axs[2].set_title('Blue Channel')
# Show the plot
plt.show()
```

**matplotlib.subplots().set_title()** : It is a function that sets the title of a subplot. The function takes a single argument: the title text.
